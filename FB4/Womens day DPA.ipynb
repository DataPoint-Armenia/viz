```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
womenmen = pd.read_csv("women_men_emp.csv")
```

```python
womenmen.groupby('Country Name')['Survey Year'].unique()
```

```python
womenmen2014 = womenmen.loc[womenmen['Survey Year'] > 2013.0]
```

```python
womenmen2014['Country Name'].unique()
```

```python
countries = ['Afghanistan', 'Armenia', 'India', 'Tajikistan', 'Ethiopia']
```

```python
womenmen2014country = womenmen2014.loc[womenmen2014['Country Name'].isin(countries)]
```

```python
nanval = womenmen2014country['Value'][0]
nanval
```

```python
womenmen2014countrynan = womenmen2014country.loc[womenmen2014country['Value'].notna()]
```

```python
womenmen2014countrynan['Indicator'][1].find('dfs')
```

```python
gender = ['Female' if x.find('women') > 0 else 'Male' for x in womenmen2014countrynan['Indicator']]
```

```python
womenmen2014countrynan['Sex'] = gender
```

```python
womenmen2014countrynan
```

```python
indicator = [x.split(' ', 4)[4] for x in womenmen2014countrynan['Indicator']]
indicator_inkind_service = ['for service only' if x == 'for in-kind payment only' else x for x in indicator]
indicator_inkind = [x if x != 'for cash and in-kind payment' else '' for x in indicator_inkind_service]
womenmen2014countrynan['Indicator'] = indicator_inkind_service
womenmen2014countrynan = womenmen2014countrynan.loc[womenmen2014countrynan['Indicator'].isin(indicator_inkind)]
```

```python
wmarmenia = womenmen2014countrynan.loc[womenmen2014countrynan['Country Name'] == "Armenia"]
wmafghan = womenmen2014countrynan.loc[womenmen2014countrynan['Country Name'] == "Afghanistan"]
wmindia = womenmen2014countrynan.loc[womenmen2014countrynan['Country Name'] == "India"]
wmethiopia = womenmen2014countrynan.loc[womenmen2014countrynan['Country Name'] == "Ethiopia"]
```

```python
wmarmenia = wmarmenia[['Indicator', 'Value', 'Sex']].drop_duplicates()
wmafghan = wmafghan[['Indicator', 'Value', 'Sex']].drop_duplicates()
wmindia = wmindia[['Indicator', 'Value', 'Sex']].drop_duplicates()
wmethiopia = wmethiopia[['Indicator', 'Value', 'Sex']].drop_duplicates()
```

```python
def plotting(country, dataset):
    sns.set_palette('RdYlBu', 9)
    sns.catplot(x="Indicator", y="Value", data=dataset, kind="bar", hue = 'Sex', height=10, aspect=.9, legend = False)
    sns.set(font_scale=1.5)
    plt.title("Married Male vs. Female Earning Type in 2015-2016: " + country)
    plt.ylim(0, 90)
    plt.ylabel('Percentage', fontsize = 18)
    plt.xlabel('Indicator', fontsize = 18)
    plt.legend(loc = 'upper right')
    plt.savefig(country + '.png', bbox_inches = 'tight', pad_inches = 1)
```
