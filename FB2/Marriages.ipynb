# Where and When did Armenians find Love the Most <3 ?

```python
### Reading the dataset
### Sournce: armstat.org
marriages = pd.read_csv('mrg.csv').loc[23:252, :]
marriages
```

```python
### converting the Date column into int 
marriages.Date = marriages.Date.astype(int)
```

```python
plt.figure(figsize = (16,10))
sns.lineplot(x= 'Date', y = 'Marriages, rate, per 1000 population', hue = 'State', data = marriages)
plt.xlabel('Date', fontsize=20)
plt.ylabel('Marriages rate, per 1000 population', fontsize=20)
plt.legend(loc=2, prop={'size': 12})
plt.show()
```

```python
plt.figure(figsize = (16,10))
sns.violinplot(x="Date", y='Marriages, rate, per 1000 population', data = marriages)
plt.xlabel('Date', fontsize=20)
plt.ylabel('Marriages rate, per 1000 population', fontsize=20)
plt.show()
```
