```python
### Importing Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from urllib.request import urlopen 
import re
import os
```

# Anush by H. Tumanyan and Natural Language Processing

```python
### Reading the text file
### Source: Wikisource.org
text_file = open("/Users/mananahakobyan/viz/FB1/anush.txt")
file_content = text_file.read()
print(file_content)
text_file.close()
```

```python
### Data Cleaning
file_content=file_content.replace('\n',' ')
```

```python
### More Data Cleaning: splitting into chapters
anush_1 = file_content.split('Chapter')
```

```python
### Reading the text into DataFrame
### where each row represents a chapter
anush_df1 = pd.DataFrame({'Chapter':anush_1})
anush_df1.head() #showing the first 5 rows
```

```python
### Getting the counts of the words
### The zero-th chapter is the prelude.

df = pd.DataFrame({
   'col1': anush_df1['Chapter'].str.count(r'(մեռ)|(մահ)|(ախ)|(վախ)|(Ախ)|(Վախ)|(նետ)'),
    'col2': np.arange(31)-1 ### chapter number
   
})
df
df1 = pd.DataFrame({
   'col1': anush_df1['Chapter'].str.count(r'(ողբ)|(Ողբ)'),
   'col2': np.arange(31)-1
})
df.head()
```

## Lineplots of Word Frequencies per Chapter

```python
# sns.set()
plt.rcParams.update({"figure.figsize" : (16, 8),
                     "axes.facecolor" : "white",
                     "axes.edgecolor":  "black"})

plt.figure(figsize=(16, 8))
plt.grid(b=None)
sns.lineplot(x='col2',y='col1',data=df,linewidth=5.5)

plt.xlabel('Chapter Number')
plt.ylabel('Frequency of Specified Words*')
plt.show()
```

```python
plt.rcParams.update({"figure.figsize" : (16, 8),
                     "axes.facecolor" : "white",
                     "axes.edgecolor":  "black"})
plt.figure(figsize=(16, 8))
plt.yticks(np.arange(0, 2, step=1))
sns.lineplot(x='col2',y='col1',data=df1,linewidth=5.5)

plt.xlabel('Chapter Number')
plt.ylabel('Frequency of Specified Words*')
plt.show()
```
